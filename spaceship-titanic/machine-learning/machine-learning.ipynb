DATA PREPARATION

```python
df = df.drop(['Name'], axis=1)
df.head()
```

```python
df[['GroupNumber', 'PassengerNumber']] = df['PassengerId'].str.split('_', expand=True)
df = df.drop(['PassengerId'], axis=1)
df = df.drop(['PassengerNumber'], axis=1)

df['GroupNumber'].astype(int)

df.head()
```

```python
df.loc[df['Cabin'].isnull()]
```

```python
def fill_cabin(df):
    # Grouping by 'GroupNumber'
    grouped = df.groupby('GroupNumber')
    
    # Iterate over each group
    for _, group in grouped:
        # Find non-null cabin values
        cabin_values = group['Cabin'].dropna().unique()
        if len(cabin_values) > 1:
            # If there are multiple non-null cabin values
            # Fill null values with the first non-null cabin value
            df.loc[group.index, 'Cabin'] = cabin_values[0]
            
fill_cabin(df)
```

```python
df.loc[df['Cabin'].isnull()]
```

```python
df = df.drop(['GroupNumber'], axis=1)
df.head()
```

```python
df[['VIP', 'CryoSleep', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'RoomService']] = df[['VIP', 'CryoSleep', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'RoomService']].fillna(value=0)
df[['HomePlanet', 'Destination']] = df[['HomePlanet', 'Destination']].fillna(value='Unknown')
df[['Cabin']] = df[['Cabin']].fillna(value='X/X/X')
df.isnull().sum().sort_values(ascending=False)
```

```python
df['Age'] = df['Age'].fillna(value=df['Age'].mean())
df.isnull().sum().sort_values(ascending=False)
```

```python
df['Transported'] = df['Transported'].astype(int)
df['VIP'] = df['VIP'].astype(int)
df['CryoSleep'] = df['CryoSleep'].astype(int)

df.head()
```

```python
df[["Deck", "Cabin_num", "Side"]] = df["Cabin"].str.split("/", expand=True)
df = df.drop(['Cabin'], axis=1)
df = df.drop(['Cabin_num'], axis=1)

df.head()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

object_columns = ['HomePlanet', 'Destination', 'Deck', 'Side'] 

for col in df[object_columns]:
    enc = label_encoder.fit(df[col])
    encoded = label_encoder.transform(df[col])
    df[col] = encoded
    print(enc.classes_)
    print(enc.transform(enc.classes_))
```

```python
df.head()
```

```python
X = df.drop(['Transported'], axis=1)
y = df['Transported']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(max_depth=2, random_state=0)
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_test, y_test)
```

```python
from sklearn.metrics import f1_score

y_pred = rfc.predict(X_test)

f1 = f1_score(y_test, y_pred)
print(f1)
```
