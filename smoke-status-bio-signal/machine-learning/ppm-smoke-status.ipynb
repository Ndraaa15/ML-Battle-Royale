PREPROCESSING

```python
df = df.drop('id', axis=1)
```

```python
df.loc[df['age'] == 0].ffill(inplace=True)
```

```python
X = df.drop(['id', 'smoking'], axis=1)
y = df['smoking']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
import xgboost as xgb

xgboost = xgb.XGBClassifier()
xgboost.fit(X_train, y_train)
```

```python
from sklearn.metrics import f1_score

y_pred = xgboost.predict(X_test)

f1 = f1_score(y_test, y_pred)
print(f1)
```

```python
df_test = pd.read_csv('../dataset/ppm-smoke-status-test.csv')
df_test.head()
```

```python
df_test.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

df_test_scaled = scaler.fit_transform(df_test.drop('id', axis=1))
```

```python
xgbproba = xgboost.predict_proba(df_test_scaled)[:, 1] 
```

```python
df_result = pd.DataFrame({
    'id': df_test['id'],
    'smoking': xgbproba
})
```

```python
df_result.to_csv('../dataset/ppm-smoke-status-result.csv', index=False)
```
