```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore')
ohe_attributes = ['attribute_0', 'attribute_1']
ohe_output = ['ohe_a_7', 'ohe_a_6', 'ohe_a_8']
ohe = OneHotEncoder(categories=[['material_5', 'material_7'],['material_5', 'material_6', 'material_8']],
                    drop='first', sparse=False, handle_unknown='ignore')
ohe.fit(df[ohe_attributes])

df[ohe_output] = ohe.transform(df[ohe_attributes])

df = df.drop(['attribute_0', 'attribute_1'], axis=1)
```

```python
prod_code_encoding = {'A': 1, 'B':2, 'C':3, 'D': 4, 'E': 5, 'F':6, 'G': 7, 'H':8, 'I': 9}

df.product_code = [prod_code_encoding[val] for val in df.product_code]
```

```python
from sklearn.impute import SimpleImputer

features = ['loading', 'measurement_3', 'measurement_4', 
            'measurement_5', 'measurement_6', 'measurement_7', 
            'measurement_8', 'measurement_9', 'measurement_10', 
            'measurement_11', 'measurement_12', 'measurement_13', 
            'measurement_14', 'measurement_15', 'measurement_16', 
            'measurement_17']

imputer = SimpleImputer(strategy="mean")
imputer.fit(df[features])

df[features] = imputer.transform(df[features])
```

CREATE MODEL

```python
X = df.drop('failure', axis=1)
y = df['failure']
```

```python
from imblearn.under_sampling import NearMiss

# define the undersampling method
undersample = NearMiss(n_neighbors=1)

# transform the dataset
X, y = undersample.fit_resample(X, y)
```

```python
# from sklearn.preprocessing import StandardScaler

# scaler = StandardScaler()
# X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.preprocessing import PowerTransformer

scaler = PowerTransformer()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, index=X.index, columns=X.columns)
X_scaled_df.head()
```
