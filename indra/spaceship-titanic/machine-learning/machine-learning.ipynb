```python
df_test[['GroupNumber', 'PassengerNumber']] = df_test['PassengerId'].str.split('_', expand=True)
df_test = df_test.drop(['PassengerNumber'], axis=1)
```

```python
df_test.loc[df_test['Cabin'].isnull()]
```

```python
# def fill_cabin(df_test):
#     # Grouping by 'GroupNumber'
#     grouped = df_test.groupby('GroupNumber')
    
#     # Iterate over each group
#     for _, group in grouped:
#         # Find non-null cabin values
#         cabin_values = group['Cabin'].dropna().unique()
#         if len(cabin_values) > 1:
#             # If there are multiple non-null cabin values
#             # Fill null values with the first non-null cabin value
#             df_test.loc[group.index, 'Cabin'] = cabin_values[0]
            
# fill_cabin(df_test)
```

```python
df_test = df_test.drop(['Name'], axis=1)
df_test[['VIP', 'CryoSleep', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'RoomService']] = df_test[['VIP', 'CryoSleep', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'RoomService']].fillna(value=0)
df_test[['HomePlanet', 'Destination']] = df_test[['HomePlanet', 'Destination']].fillna(value='Unknown')
df_test['Cabin'] = df_test['Cabin'].fillna(value='X/X/X')
df_test.head()
```

```python
df_test[["Deck", "Cabin_num", "Side"]] = df_test["Cabin"].str.split("/", expand=True)
df_test = df_test.drop(['Cabin'], axis=1)
df_test = df_test.drop(['Cabin_num'], axis=1)

df_test.head()
```

```python
# Mengganti nilai 'X' di kolom 'Deck' dengan modus dari kolom 'Deck'
df_test['Deck'].replace('X', df_deck_side['Deck'].mode()[0], inplace=True)

# Mengganti nilai 'X' di kolom 'Side' dengan modus dari kolom 'Deck'
df_test['Side'].replace('X', df_deck_side['Side'].mode()[0], inplace=True)
```

```python
df_test['Age'] = df_test['Age'].fillna(value=df_test['Age'].mean())
df_test.isnull().sum().sort_values(ascending=False)
```

```python
df_test['VIP'] = df_test['VIP'].astype(int)
df_test['CryoSleep'] = df_test['CryoSleep'].astype(int)

df_test.head()
```

```python
passenger_id = df_test['PassengerId']
df_test = df_test.drop(['PassengerId'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

object_columns = ['HomePlanet', 'Destination', 'Deck', 'Side'] 

for col in df_test[object_columns]:
    enc = label_encoder.fit(df_test[col])
    encoded = label_encoder.transform(df_test[col])
    df_test[col] = encoded
    print(enc.classes_)
    print(enc.transform(enc.classes_))
```

```python
df_test.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_test_scaled = scaler.fit_transform(df_test)
```

```python
df_result = pd.DataFrame({  
  'PassengerId': passenger_id,
  'Transported': rfc.predict(df_test_scaled).astype(bool)
}
)

df_result.head()
```
